```python
# Tells matplotlib to display images inline instead of a new window
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('ggplot')


```

```python
# read data into memory
googprice = np.genfromtxt("GooglePrice.csv", delimiter = ",")
# get month and sales values
day = googprice[:,0].astype(int)
price = googprice[:,1]
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(day,price, 'b')
plt.xlabel("day")
plt.ylabel("price")
plt.title("Google Stock Price")
plt.legend(['Google Stock Price'])
plt.show()
```

```python
diffprice = np.diff(price)
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(day[0:251],diffprice[0:251], 'b')
plt.xlabel("day")
plt.ylabel("difference (price)")
plt.title("Differenced Google Stock Price")
plt.legend(['Differenced Google Stock Price'])
plt.show()
```

```python
# read data into memory
bmw = np.genfromtxt("bmw_sales.csv", delimiter = ",")
# get month and sales values
month = bmw[:,0].astype(int)
salesbmw = bmw[:,1]
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(month,salesbmw, 'r')
plt.xlabel("month")
plt.ylabel("Sales")
plt.title("Monthly Sales of BMWs")
plt.legend(['Monthly Sales of BMWs'])
plt.show()
```

```python
sdsalesbmw=[0]*73;
for i in range(12,71):
 sdsalesbmw[i] =salesbmw[i]- salesbmw[i-12]
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(month[12:71],sdsalesbmw[12:71], 'r')
plt.xlabel("month")
plt.ylabel("Deasonalized Sales")
plt.title("Deasonalized Monthly Sales of BMWs")
plt.legend(['Deasonalized Monthly Sales of BMWs'])
plt.show()
```

```python
dsdsalesbmw=np.diff(sdsalesbmw)
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(month[12:71],dsdsalesbmw[12:71], 'r')
plt.xlabel("month")
plt.ylabel("Differenced and Deasonalized Sales")
plt.title(" Differenced  Deasonalized Monthly Sales of BMWs")
plt.legend(['Differenced  and Deasonalized Monthly Sales of BMWs'])
plt.show()
```

```python
# read data into memory
covitaly = np.genfromtxt("covid_italy.csv", delimiter = ",")
# get month and sales values
day = covitaly [:,0].astype(int)
casesit = covitaly [:,1]
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(day,casesit, 'y')
plt.xlabel("day")
plt.ylabel("New Cases")
plt.title("Daily New Covid Cases")
plt.legend(['Daily New Covid Cases'])
plt.show()
```

```python
lcasesit=np.log(casesit)
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(day,lcasesit, 'y')
plt.xlabel("day")
plt.ylabel("Log(New Cases)")
plt.title("Log(Daily New Covid Cases)")
plt.legend(['Log(Daily New Covid Cases)'])
plt.show()
```

```python
#K=X.shape[0]
plt.figure(figsize = (16, 8))
plt.plot(day,dlcasesit, 'y')
plt.xlabel("day")
plt.ylabel("Log(New Cases)")
plt.title("Log(Daily New Covid Cases)")
plt.legend(['Log(Daily New Covid Cases)'])
plt.show()
```
