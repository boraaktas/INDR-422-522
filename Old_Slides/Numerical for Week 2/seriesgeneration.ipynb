```python
# Tells matplotlib to display images inline instead of a new window
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
plt.style.use('ggplot')
# import the random module
import random
  
```

```python
# Generate a stationary series with desired level (c) and standard deviation (sigma)


c=100; sigma=15;
y=[0]*101;
for i in range(1,100):
 y[i] =c + sigma*random.normalvariate(0, 1)
```

```python
plt.plot(y[1:100])
```

```python
# Generate a series with linear trend with desired level (c), slope(b) and standard deviation (sigma)

c=100; sigma=15; b=2; 
ytr=[0]*101;
for i in range(1,100):
 ytr[i] =c + b*i+ sigma*random.normalvariate(0, 1)
```

```python
plt.plot(ytr[1:100])
```

```python
# Generate a series with quadratic trend and with desired  parameters

c=100; sigma=25; b=2; a=0.05; 
yqr=[0]*101;
for i in range(1,100): 
 yqr[i] =c + b*i+ a*i*i+ sigma*random.normalvariate(0, 1)
```

```python
plt.plot(yqr[1:100])
```

```python
# Generate a seasonal series for a four period cycle with seasonal levels c(i) 
ssc=[0]*4; # seasonal level vector
ssc[0]=100; ssc[1]=150; ssc[2]=200; ssc[3]=120; sigma=15; b=2;  # seasonal levels and other parameters
yss=[0]*101;
for i in range(1,100):
 yss[i] =ssc[i % 4] + sigma*random.normalvariate(0, 1) # ssc[i % 4] takes care of the modulo operation ssc[i % 4] = ssc[i(mod 4)]
```

```python
plt.plot(yss[1:36])
```
